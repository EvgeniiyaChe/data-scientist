```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
```

```python
n_samples = 1000

years_car = np.random.choice(8, n_samples) + 1
rate = np.random.choice(25, n_samples) + 5
power = np.random.choice(250, n_samples) + 70
cost_car = np.random.choice(300000, n_samples) + 30000

tax = years_car * rate * power

data = pd.DataFrame({'years_car': years_car, 'rate': rate, 'power': power, 'tax': tax, 'cost_car': cost_car})
data.head(10)
```

```python
from sklearn.metrics import mean_absolute_error

X = data[['years_car', 'rate', 'power']]
y = data['tax']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['years_car', 'rate', 'power']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```

```python
X = data[['rate', 'power']]
y = data['tax']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['rate', 'power']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```

```python
# Создаем новый признак (считаем повышающий коэффициент)
data['coefficient'] = data['cost_car'] / data['years_car'] * data['power'] * data['rate'] / 100
data.head(10)
```

```python
X = data[['coefficient']]
y = data['tax']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['coefficient']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```
